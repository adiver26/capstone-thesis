```python
import pandas as pd
import numpy as np
import os
import xml.etree.cElementTree as et
```

```python
df = pd.read_csv("dataset.csv")
```

```python
path = "C:\\Users\\Deepank\\Downloads\\XML Files - BBox"
xml_list = []

for root, dirs, files in os.walk(path):
    for file in files:
        if(file.endswith(".xml")):
            print(os.path.join(root,file))
            xml_list.append(os.path.join(root,file))
```

```python
xmin = []
xmax = []
ymin = []
ymax = []
```

```python
for i in enumerate(xml_list):
    if i[0]==120:
        for n in range(60):
            xmin.append("?")
            xmax.append("?")
            ymin.append("?")
            ymax.append("?")
    tree = et.parse(i[1])
    root = tree.getroot()
    for xmn in root.iter('xmin'):
        xmin.append(xmn.text)
    for xmx in root.iter('xmax'):
        xmax.append(xmx.text)
    for ymn in root.iter('ymin'):
        ymin.append(ymn.text)
    for ymx in root.iter('ymax'):
        ymax.append(ymx.text)
```

```python
df['Xmin'] = xmin
df['Ymin'] = ymin
df['Xmax'] = xmax
df['Ymax'] = ymax
```

```python
df.to_csv("dataset-New.csv")
```
