```python
import tensorflow as tf
from keras import models
from keras.models import load_model
from keras import optimizers
import keras.backend as K
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.layers.core import Dense
```

```python
data = data[data.Disease!="Healthy leaf"]
data.reset_index(drop=True, inplace=True)
```

```python
data.Xmax = data.Xmax.astype("int64")
data.Ymax = data.Ymax.astype("int64")
data.Xmin = data.Xmin.astype("int64")
data.Ymin = data.Ymin.astype("int64")
```

```python
data.info()
```

```python
# Splitting the dataset into train and test

ratio = int(0.8 * data.shape[0])
indices = np.random.permutation(data.shape[0])
training_idx, test_idx = indices[:ratio], indices[ratio:]
train, test = data.iloc[training_idx,:], data.iloc[test_idx,:]

train.reset_index(drop=True, inplace=True)
test.reset_index(drop=True, inplace=True)
```

```python
train.info()
```

```python
test.info()
```

```python
# Function to normalize bounding box coordinates (PASCAL VOC format)

def normalize(box):             # box -> (xmin, ymin, xmax, ymax)
    x1 = box[0] / 224
    y1 = box[1] / 224
    x2 = box[2] / 224
    y2 = box[3] / 224
    return (x1, y1, x2, y2)
```

```python
# Preparing training dataset labels for input

labels = np.zeros(shape = (len(train), 4))

for i in range(len(train)):
    box = (train['Xmin'][i], train['Ymin'][i], train['Xmax'][i], train['Ymax'][i])
    bb = normalize(box) 
    labels[i] = [bb[0], bb[1], bb[2], bb[3]]
```

```python
# Preparing test dataset labels for evaluation

val_labels = np.zeros(shape = (len(test), 4))

for i in range(len(test)):
    box = (test['Xmin'][i], test['Ymin'][i], test['Xmax'][i], test['Ymax'][i])
    bb = normalize(box) 
    val_labels[i] = [bb[0], bb[1], bb[2], bb[3]]
```

```python
# Loading the pre-trained model 

model = VGG16(weights='imagenet', include_top=False)
```

```python
# Variable initialisation for model training and evaluation

train_feature = np.zeros(shape = (len(train), 7, 7, 512)) # Last layer of VGG16 has a shape of 7 x 7 x 512
test_feature = np.zeros(shape = (len(test), 7, 7, 512))
```

```python
# Feature extraction of training set images

for i in range(len(train)):
    img_path = train.Images[i]
    img = cv2.imread(img_path)
    img_data = cv2.resize(img, (224,224))
    img_data = np.expand_dims(img_data, axis=0)
    img_data = preprocess_input(img_data)
    vgg16_feature = model.predict(img_data)
    train_feature[i] = vgg16_feature
```

```python
# Feature extraction of test set images

for i in range(len(test)):
    img_path = test.Images[i]
    img = cv2.imread(img_path)
    img_data = cv2.resize(img, (224,224))
    img_data = np.expand_dims(img_data, axis=0)
    img_data = preprocess_input(img_data)
    vgg16_feature = model.predict(img_data)
    test_feature[i] = vgg16_feature
```

```python
train_features_vec = np.reshape(train_feature, (len(train), 7 * 7 * 512))
print("Train features of the image: {}".format(train_features_vec.shape))

test_features_vec = np.reshape(test_feature, (len(test), 7 * 7 * 512))
print("Train features of the image: {}".format(test_features_vec.shape))
```

```python
# Defining functions for calculating the model loss and accuracy

def calculate_iou(target, pred):
    xA = K.maximum(target[:,0], pred[:,0])
    yA = K.maximum(target[:,1], pred[:,1])
    xB = K.minimum(target[:,2], pred[:,2])
    yB = K.minimum(target[:,3], pred[:,3])
    interArea = K.maximum(0.0, xB-xA+1) * K.maximum(0.0, yB-yA+1)
    boxAarea = (target[:,2]-target[:,0]+1) * (target[:,3]-target[:,1]+1)
    boxBarea = (pred[:,2]-pred[:,0]+1) * (pred[:,3]-pred[:,1]+1)
    iou = interArea / float(boxAarea + boxBarea - interArea)
    return iou

def custom_loss(y_true, y_pred):
    mse = tf.losses.mean_squared_error(y_true, y_pred)
    iou = calculate_iou(y_true , y_pred)
    return mse + (1-iou)

def iou_metric(y_true, y_pred):
    return calculate_iou(y_true, y_pred)
```

```python
# Modifying and re-using the saved new neural network model

new_model = load_model("mymodel.h5")
new_model.pop()
new_model.add(Dense(4, activation='linear', name="last_layer"))
```

```python
new_model.compile(optimizer=optimizers.Adam(0.0001), loss=custom_loss, metrics=[iou_metric])
```

```python
# Model training and validation

history = new_model.fit(train_features_vec, labels, validation_data=(test_features_vec, val_labels), epochs=400)
```

```python
# Visualising model performance

loss = history.history['loss']
val_loss = history.history['val_loss']

acc = history.history['iou_metric']
val_acc = history.history['val_iou_metric']

plt.figure(figsize=(10,15))
plt.subplot(2,1,1)
plt.plot(loss , linewidth=3 ,label='train loss')
plt.plot(val_loss , linewidth=3, label='val loss')
plt.xlabel('epochs')
plt.ylabel('loss / val_loss')
plt.legend()

plt.subplot(2,1,2)
plt.plot(acc , linewidth=3 ,label='train acc')
plt.plot(val_acc , linewidth=3, label='val acc')
plt.xlabel('epochs')
plt.ylabel('Accuracy / Val_Accuracy')
plt.legend()
```

```python
# Understanding model performance

predict = new_model.predict(test_features_vec)
predbox = predict[...,0:4] * 224         # denormalisation
groundtruth = val_labels[...,0:4] * 224
```

```python
# Function to find IOU (to be used while plotting)

def bb_iou(boxA, boxB):
    # determine the (x, y)-coordinates of the intersection rectangle
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    
    # compute the area of intersection rectangle
    interArea = max(0, xB - xA + 1) * max(0, yB - yA + 1)
    
    # compute the area of both the prediction and ground-truth
    boxAArea = (boxA[2] - boxA[0] + 1) * (boxA[3] - boxA[1] + 1)
    boxBArea = (boxB[2] - boxB[0] + 1) * (boxB[3] - boxB[1] + 1)
    
    # compute the intersection over union by taking the intersection
    # area and dividing it by the sum of prediction + ground-truth
    # areas - the interesection area
    iou = interArea / float(boxAArea + boxBArea - interArea)
    return iou
```

```python
# Visualising relatively better detection predictions

for i in range(len(test)):
    if bb_iou(groundtruth[i], predbox[i]) > 0.35:
        data = im.imread(test.Images[i])

        start_point = (int(predbox[i][0]), int(predbox[i][1]))               # (xmin, ymin)
        end_point = (int(predbox[i][2]), int(predbox[i][3]))                 # (xmax, ymax)
        a_start_point = (int(groundtruth[i][0]), int(groundtruth[i][1]))     # (xmin, ymin)
        a_end_point = (int(groundtruth[i][2]), int(groundtruth[i][3]))    

        color = (255, 0, 0)       # RGB
        a_color = (0, 0, 255)                       
        
        if data.shape[0] < 300:
            thickness = 3
        else:
            thickness = 9
        
        data = cv2.rectangle(data, start_point, end_point, color, thickness)
        data = cv2.rectangle(data, a_start_point, a_end_point, a_color, thickness)
        
        plt.imshow(data)
        plt.show()
```

```python
# Viewing the mis-detections

count = 0

for i in range(len(test)):
    if bb_iou(groundtruth[i], predbox[i]) == 0 and count < 5:
        data = im.imread(test.Images[i])

        start_point = (int(predbox[i][0]), int(predbox[i][1]))               # (xmin, ymin)
        end_point = (int(predbox[i][2]), int(predbox[i][3]))                 # (xmax, ymax)
        a_start_point = (int(groundtruth[i][0]), int(groundtruth[i][1]))     # (xmin, ymin)
        a_end_point = (int(groundtruth[i][2]), int(groundtruth[i][3]))    

        color = (255, 0, 0)       # RGB
        a_color = (0, 0, 255)                       
        
        if data.shape[0] < 250:
            thickness = 3
        else:
            thickness = 9
        
        data = cv2.rectangle(data, start_point, end_point, color, thickness)
        data = cv2.rectangle(data, a_start_point, a_end_point, a_color, thickness)
        count += 1
        
        plt.imshow(data)
        plt.show()
```
